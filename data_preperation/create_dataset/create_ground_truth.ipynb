```python
path_shp = "/home/philipp/Data/edin_diss/GIS_ground_truth/df_2020_2/gis_df_wwie_2020_177.shp"
wo_geo = gpd.read_file(path_shp)
```

```python
wo_geo.head()
```

```python
# std
# conditions 0 = void, 1 = thinning (ac1+2), 3 = no thinning, 4 = else
conditions = [
    (wo_geo['typ'] == 0),
    (wo_geo['typ'] == 1),
    ((wo_geo['typ'] == 2) | (wo_geo['typ'] == 4)),
    (wo_geo['typ'] == 3)]

choices = [0, 1, 2, 3]

wo_geo['typ_std'] = np.select(conditions, choices, default=0)
```

```python
# add
# conditions 0 = void, 1 = thinning, 5 = just thinned, 3 = no thinning, 4 = else
conditions = [
    (wo_geo['typ'] == 0),
    (wo_geo['typ'] == 1),
    (wo_geo['typ'] == 4),
    (wo_geo['typ'] == 2),
    (wo_geo['typ'] == 3)]

choices = [0, 1, 2, 3, 4]

wo_geo['typ_add'] = np.select(conditions, choices, default=0)
```

```python
# acutness 1 & 2
# conditions 0 = void, 1 = thinning (ac1), 2 = thinning (ac2), 3 = no thinning, 4 = else
conditions = [
    (wo_geo['typ'] == 0),
    ((wo_geo['typ'] == 1) & (wo_geo['dr'] == 1)),
    ((wo_geo['typ'] == 1) & ((wo_geo['dr'] == 0) | (wo_geo['dr'] > 1))),
    ((wo_geo['typ'] == 2) | (wo_geo['typ'] == 4)),
    (wo_geo['typ'] == 3)]

choices = [0, 1, 2, 3, 4]

wo_geo['typ_ac12'] = np.select(conditions, choices, default=0)
```

```python
# acutness 1
# conditions 0 = void, 1 = thinning (ac1), 3 = no thinning, 4 = else
conditions = [
    (wo_geo['typ'] == 0), \
    ((wo_geo['typ'] == 1) & (wo_geo['dr'] == 1)), \
    (((wo_geo['typ'] == 1) & (wo_geo['dr'] != 1)) | \
     (wo_geo['typ'] == 2) | \
     (wo_geo['typ'] == 4)), \
    (wo_geo['typ'] == 3)]

choices = [0, 1, 2, 3]

wo_geo['typ_ac1'] = np.select(conditions, choices, default=0)
```

```python
############
# set mask
# economy = I; ww_sw = W; age < uz -20
wo_geo['typ_mask'] = 0
wo_geo.loc[(wo_geo['ww_sw']=='W') & \
           (wo_geo['economy']=='I') & \
           (wo_geo['age']<=wo_geo['uz']-20), 'typ_mask'] = \
wo_geo.loc[(wo_geo['ww_sw']=='W') & \
           (wo_geo['economy']=='I') & \
           (wo_geo['age']<=wo_geo['uz']-20), 'typ']

# print shape
wo_geo.loc[(wo_geo['ww_sw']=='W') & \
           (wo_geo['economy']=='I') & \
           (wo_geo['age']<=wo_geo['uz']-20),:].shape
```

```python
## std_mask
# conditions 0 = void, 1 = thinning (ac1+2), 3 = no thinning, 4 = else
conditions = [
    (wo_geo['typ_mask'] == 0),
    (wo_geo['typ_mask'] == 1),
    ((wo_geo['typ_mask'] == 2) | (wo_geo['typ_mask'] == 4)),
    (wo_geo['typ_mask'] == 3)]

choices = [0, 1, 2, 3]

wo_geo['typ_std_m'] = np.select(conditions, choices, default=0)
```

```python
## acutness 1 & 2 mask
# conditions 0 = void, 1 = thinning (ac1), 2 = thinning (ac2), 3 = no thinning, 4 = else
conditions = [
    (wo_geo['typ_mask'] == 0),
    ((wo_geo['typ_mask'] == 1) & (wo_geo['dr'] == 1)),
    ((wo_geo['typ_mask'] == 1) & ((wo_geo['dr'] == 0) | (wo_geo['dr'] > 1))),
    ((wo_geo['typ_mask'] == 2) | (wo_geo['typ_mask'] == 4)),
    (wo_geo['typ_mask'] == 3)]

choices = [0, 1, 2, 3, 4]

wo_geo['typ_ac12_m'] = np.select(conditions, choices, default=0)
```

```python
# acutness 1
# conditions 0 = void, 1 = thinning (ac1), 3 = no thinning, 4 = else
conditions = [
    (wo_geo['typ_mask'] == 0),
    ((wo_geo['typ_mask'] == 1) & (wo_geo['dr'] == 1)),
    (((wo_geo['typ_mask'] == 1) & (wo_geo['dr'] != 1)) | \
     (wo_geo['typ_mask'] == 2) | \
     (wo_geo['typ_mask'] == 4)),
    (wo_geo['typ_mask'] == 3)]

choices = [0, 1, 2, 3]

wo_geo['typ_ac1_m'] = np.select(conditions, choices, default=0)
```

```python
wo_geo.to_file("/home/philipp/Data/edin_diss/GIS_ground_truth/df_2020_2/gis_df_wwie_2020_177_adj_0801.shp")
```
